## "Mutating" Columns

```python
demographic_data.rename(columns = {"income_per_person":"income_gdp_ppp_dollars"}) # old name : new name
```

```python
demographic_data["years_since_1900"] = demographic_data["year"] - 1900 # make a new column and define it
```

```python
demographic_data
```

```python
demographic_data.loc[:, "income_per_person"] = demographic_data["income_per_person"] * 0.8 # changes existing column
```

```python
demographic_data.head()
```

```python
demographic_data.loc[:, "life_expectancy"] = demographic_data["life_expectancy"].map(round)
```

```python
demographic_data.head()
```

```python
occupation_prestige.loc[:, "income"] = occupation_prestige["income"]/1000
```

```python
occupation_prestige.head()
```

## Grouping and Summarising

```python
demographic_data.loc[:, "life_expectancy"].max()
```

```python
demographic_data.loc[demographic_data["country"] == "Afghanistan", "life_expectancy"].mean()
```

```python
demographic_data.groupby("country")["income_per_person"].mean().reset_index(name = "mean_income")
```

## Missing Values

```python
dem_missin = pd.read_csv("data/life_expectancy_and_income_missing.csv")
```

```python
dem_missin.head()
```

```python
dem_missin.info()
```

```python
dem_missin.isna().sum() # counting NA's
```

```python
dem_missin.fillna(0)
```

```python
dem_missin.fillna({"life_expectancy": dem_missin.groupby("country")["life_expectancy"].transform("mean"), 
                   "income_per_person": dem_missin["income_per_person"].mean()})
```

```python
dem_missin.dropna()
```
