```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

```python
demo_data = pd.read_csv("data/life_expectancy_and_income.csv")
demo_data_2019 = demo_data.loc[demo_data["year"] == 2019].copy()
```

```python
sns.scatterplot(x = "income_per_person",
                y = "life_expectancy",
                color = "red",
               data = demo_data_2019);
```

```python
sns.scatterplot(x = "income_per_person",
                y = "life_expectancy",
                hue = "fertility_rate",
                palette = "mako_r",
               data = demo_data_2019);
```

```python
sns.scatterplot(x = "income_per_person",
                y = "life_expectancy",
                size = "fertility_rate",
                hue = "fertility_rate",
               data = demo_data_2019);
```

```python
demo_data_2019.loc[:, "fertility_above_5"] = np.where(demo_data_2019["fertility_rate"] > 5, "yes", "no")
```

```python
sns.scatterplot(x = "income_per_person",
                y = "life_expectancy",
                hue = "fertility_above_5",
                palette = dict(yes = "green", no = "red"),
               data = demo_data_2019);
```

```python
sns.scatterplot(x = "income_per_person",
                y = "life_expectancy",
                style = "fertility_above_5",
                hue = "fertility_above_5",
                alpha = 0.3,
               data = demo_data_2019);
```

```python
occupation_prestige = pd.read_csv("data/prestige_occupation.csv")
occupation_prestige.head()
```

```python
sns.scatterplot(x = "education",
                y = "prestige",
                hue = "type",
                palette = "viridis",
                alpha = 0.7,
               data = occupation_prestige)
plt.legend(bbox_to_anchor = (1, 1));
```

```python
sns.scatterplot(x = "education",
                y = "prestige",
                hue = "income",
                palette = "viridis",
                alpha = 0.7,
               data = occupation_prestige)
plt.legend(bbox_to_anchor = (1, 1));
```

```python
sns.set_style("dark") # set a theme for a chart

sns.scatterplot(x = "education",
                y = "prestige",
                hue = "type",
                palette = "viridis",
                alpha = 0.7,
               data = occupation_prestige)
plt.legend(bbox_to_anchor = (1, 1));
```
