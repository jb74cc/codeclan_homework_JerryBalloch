# Plotting in Python

```python
import pandas as pd
```

```python
demo_data = pd.read_csv("data/life_expectancy_and_income.csv")
```

```python
demo_data
```

```python
demo_data_2019 = demo_data.loc[demo_data["year"] == 2019].copy()
```

```python
demo_data_2019.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [5, 3]
```

```python
plt.rcParams['figure.dpi'] = 120
```

```python
plt.rcParams['font.size'] = 10
```

```python
plt.scatter(x = "income_per_person", y = "life_expectancy", data = demo_data_2019)
plt.xlabel("Income per person ($, ppp)")
plt.ylabel("Life expectancy")
plt.title("Scatterplot of Income per person vs. life expectancy");
```

```python
demo_data_2019.loc[:, "income_per_person_1000"] = demo_data_2019.loc[:, "income_per_person"]/1000
```

```python
demo_data_2019
```

```python
plt.scatter(x = "income_per_person_1000", y = "life_expectancy", data = demo_data_2019)
plt.xlabel("Income per person ($1000, ppp)")
plt.ylabel("Life expectancy")
plt.title("Scatterplot of Income per person vs. life expectancy", size = 10);
```

```python
plt.scatter(x = "income_per_person_1000", y = "life_expectancy", data = demo_data_2019)
plt.xlim(0, 10)
plt.xlabel("Income per person ($1000, ppp)")
plt.ylabel("Life expectancy")
plt.title("Scatterplot of Income per person vs. life expectancy [DETAIL]", size = 10);
```

```python
plt.scatter(x = "income_per_person_1000", y = "life_expectancy", data = demo_data_2019)
plt.yticks(50, 60, 70, 80, 90)
plt.xlim(0, 10)
plt.xlabel("Income per person ($1000, ppp)")
plt.ylabel("Life expectancy")
plt.title("Scatterplot of Income per person vs. life expectancy [DETAIL]", size = 10);
```

```python
plt.hist(x = "income_per_person", data = demo_data_2019);
```

```python
plt.boxplot(x = "income_per_person", data = demo_data_2019);
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x = "income_per_person_1000", y = "life_expectancy", data = demo_data_2019);
```

```python
sns.histplot(x = "income_per_person_1000", data = demo_data_2019);
```

```python
sns.boxplot(x = "income_per_person_1000", data = demo_data_2019);
```

```python
demo_data_japan = demo_data.loc[demo_data["country"] == "Japan"].copy()
```

```python
demo_data_japan.head()
```

```python
sns.lineplot(x = "year", y = "fertility_rate", data = demo_data_japan);
plt.title("Fertility rate over time for Japan");
```

```python
sns.barplot(x = "country", y = "life_expectancy", data = demo_data_2019.head(5));
```

```python
ordered_data = demo_data_2019.head(5).sort_values("life_expectancy").country.copy()
```

```python
ordered_data
```

```python
sns.barplot(x = "country", y = "life_expectancy", data = demo_data_2019.head(5), order = ordered_data);
```

```python
# alternative without having to create order by data
sns.barplot(x = "country", y = "life_expectancy", data = demo_data_2019.head(5).sort_values("life_expectancy"));
```

```python
last_20 = demo_data.loc[demo_data["year"] >= 2000].copy()
```

```python
last_20
```

```python
import numpy as np
```

```python
sns.barplot(x = "year",
            y = "income_per_person",
           data = last_20,
           estimator = np.mean);
```

```python
sns.barplot(x = "year",
            y = "income_per_person",
           data = last_20,
           estimator = np.mean,
           ci = False)
plt.title("Average income per person by year from 2000-2019")
plt.xticks(rotation = 72);
```

```python
sns.barplot(y = "year",
            x = "income_per_person",
           data = last_20,
           estimator = np.mean,
           ci = False,
           orient = "h").invert_yaxis()
plt.title("Average income per person by year from 2000-2019");
```
